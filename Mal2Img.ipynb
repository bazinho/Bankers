```python
import os
import glob
import optparse
import scipy.misc
import numpy as np
import array
```

```python
def GetWidth(size):
    widthtable = [(10240, 32),
                  (30720, 64),
                  (61440, 128),
                  (102400, 256),
                  (204800, 384),
                  (512000, 512),
                  (1024000, 768)]
    maxwidth = 1024
    for w in widthtable:
        if size<w[0]:
            return w[1]
    return maxwidth 
```

```python
def Mal2Img(malwaredir,imagedir):
    malwarelist = glob.glob(malwaredir+'/*.exe')
    for i,malware in enumerate(malwarelist):
        print("[%d] Converting file: %s" %(i,malware))
        size = os.path.getsize(malware)
        width = GetWidth(size)
        rem = size%width
        a = array.array("B")
        f = open(malware,'rb')
        a.fromfile(f,size-rem)
        f.close()
        g = np.reshape(a,(len(a)//width,width))
        g = np.uint8(g)
        image = os.path.basename(malware).split('.')[0]+'.png'
        scipy.misc.imsave(os.path.join(imagedir,image),g)
```

```python
datasetdir = "Datasets/Detection"
malwaredir = "Malwares"
imagedir = "Byteplots"
classes = sorted(os.listdir(os.path.join(datasetdir,malwaredir)))  # vector of strings with family names
num_samples = []
for i in range(len(classes)):
    cur_dir = os.path.join(datasetdir,malwaredir,classes[i])
    num_samples.append(len(glob.glob(os.path.join(cur_dir,'*.exe'))))
    print ("Class: %15s\tNumber of samples: %d" % (classes[i], num_samples[-1]))
    if not os.path.exists(os.path.join(datasetdir,imagedir,classes[i])):
        print("Creating dir: %s" %(os.path.join(datasetdir,imagedir,classes[i])))
        os.mkdir(os.path.join(datasetdir,imagedir,classes[i]))
    Mal2Img(os.path.join(datasetdir,malwaredir,classes[i]),os.path.join(datasetdir,imagedir,classes[i]))
total_samples = np.sum(num_samples)
```
